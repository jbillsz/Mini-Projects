```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
sin=np.sin(np.arange(100))
cos=np.cos(np.arange(100))
ax1=plt.subplot(121, facecolor='Red')
plt.plot(sin,'-o',cos,'-x')
ax2=plt.subplot(122, facecolor='yellow')
plt.scatter(sin,cos, s=1)
```

```python
dist= np.random.normal(loc=6,scale=8,size=1000)
plt.plot(dist)
```

```python
plt.figure()  
fig , ((ax1,ax2,ax3),(ax4,ax5,ax6),(ax7,ax8,ax9))= plt.subplots(3,3,sharex=True, sharey=True)
ax1.plot(sin)
```

```python
plt.bar(dist,height=[x for x in np.random.randint(1,100,1000)]) #plotted a random distribution with a random height values
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2, sharex=True,sharey=True)
ax=[ax1,ax2,ax3,ax4]
for n in range(0,len(ax)):
    sample_size=10**n+1
    sample=np.random.normal(loc=0, scale=1, size=sample_size)
    ax[n].hist(sample)
```

```python
import matplotlib.gridspec as gridspec
plt.figure()
gridspec=gridspec.GridSpec(3,3)
top=plt.subplot(gridspec[0,0])
top.plot(sin)
left= plt.subplot(gridspec[0,1])
left.scatter(cos,sin)
left_left=plt.subplot(gridspec[0,2])
plt.plot(cos)
mid_first=plt.subplot(gridspec[1,0])
mid_first.hist(np.random.normal(loc=0,scale=1,size=1000))
```
